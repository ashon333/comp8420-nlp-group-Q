# COMP8420 Assignment 3: Basic NLP Techniques
This notebook demonstrates the complete implementation and execution of **all 6 Basic NLP Techniques** in our E-commerce Intelligent System.

### 1. Preprocessing Pipeline
Fuses normalization, tokenization, stopword removal, and WordNet lemmatization.

```python
import sys
import os
sys.path.append(os.path.abspath('../'))

from src.preprocessing import TextPreprocessor
preprocessor = TextPreprocessor()
sample_text = "The SonicWave headphones are AMAZING! It works perfectly and is highly durable."
res = preprocessor.process_pipeline(sample_text)
print('Original Text:', sample_text)
print('Normalized Text:', res['normalized'])
print('Tokens:', res['tokens'])
print('Lemmatized Tokens:', res['lemmatized'])
```

### 2. POS Tagging & Aspect-Opinion Extraction
Syntactic dependency parsing to extract product aspects.

```python
from src.extraction import InfoExtractor
extractor = InfoExtractor()
aspects = extractor.extract_pos_and_aspects(sample_text)
print('Extracted aspect-opinion pairs:', aspects['aspect_pairs'])
```

### 3. Named Entity Recognition (NER)
Extracts brands, specs, and locations.

```python
ner_text = "We bought SonicWave headphones at Apple Store in Sydney for $149."
entities = extractor.extract_named_entities(ner_text)
print('Named Entities:', entities)
```

### 4. Rule-Based Information Extraction (Regex)
Pulls numerical price tags and spec formats (e.g. 15-bar, USB-C).

```python
regex_text = "The EspressoMaster is a 15-bar coffee maker with USB-C charging priced at $299.00."
specs = extractor.extract_prices_and_versions(regex_text)
print('Extracted Prices:', specs['extracted_prices'])
print('Extracted Specs:', specs['extracted_specs'])
```

### 5. Traditional ML Baseline Sentiment
TF-IDF + Logistic Regression baseline model.

```python
from src.traditional_ml import TraditionalMLClassifier
classifier = TraditionalMLClassifier()
pos_pred = classifier.predict("This machine is incredible and brews the best coffee!")
neg_pred = classifier.predict("Broke immediately. Leaked all over my counter.")
print('Positive Review prediction label (1 = Positive):', pos_pred)
print('Negative Review prediction label (0 = Negative):', neg_pred)
```

### 6. Sentence Embeddings & Cosine Search
Vectors-based search using Cosine Similarity.

```python
from src.semantic_search import SemanticSearchEngine
search_engine = SemanticSearchEngine()
corpus = [
    {'id': 1, 'description': '15-bar stainless steel espresso maker'},
    {'id': 2, 'description': 'Active noise canceling headphones'}
]
search_engine.index_corpus(corpus, text_key='description')
search_res = search_engine.search('make coffee')
print('Matched Search Item:', search_res[0]['item']['description'], '(Similarity:', search_res[0]['score'], ')')
```
